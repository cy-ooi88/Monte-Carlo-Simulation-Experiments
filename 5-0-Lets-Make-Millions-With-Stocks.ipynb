```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(5)
stock_returns = np.random.normal(loc=0.001, scale=0.05, size=1000)
```

```python
plt.plot(stock_returns.cumsum())
```

```python
avg, std = stock_returns.mean(), stock_returns.std()
avg, std
```

```python
simulated_stock_returns = np.random.normal(loc=avg, scale=std, size=1000)
plt.plot(stock_returns.cumsum())
plt.plot(simulated_stock_returns.cumsum())
```

```python
last_return_vals = []

for _ in range(1000):
    simulated_stock_returns = np.random.normal(loc=avg, scale=std, size=1000)
    plt.plot(simulated_stock_returns.cumsum(), alpha=0.05)
    last_return_vals.append(simulated_stock_returns.cumsum()[-1])
    
plt.plot(stock_returns.cumsum(), color='black', linewidth=2)
```

```python
last_return_vals = np.array(last_return_vals)
plt.hist(last_return_vals, bins=50);
print('Average: {:.1f}, Minimum: {:.1f}, Maximum: {:.1f}'.format(last_return_vals.mean(),
                                                                 last_return_vals.min(),
                                                                 last_return_vals.max()))
```
